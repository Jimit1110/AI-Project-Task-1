```python
from transformers import MT5ForConditionalGeneration, MT5Tokenizer

# Load the pre-trained model and tokenizer
model_name = "rooftopcoder/mT5_base_English_Gujrati"
model = MT5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = MT5Tokenizer.from_pretrained(model_name)

def translate_text(text, max_length=200):
    # Tokenize and generate translation
    input_ids = tokenizer.encode(text, return_tensors="pt", max_length=max_length, truncation=True)
    translated_ids = model.generate(input_ids)

    # Decode the translated text
    translated_text = tokenizer.batch_decode(translated_ids, skip_special_tokens=True)[0]

    return translated_text

# Example usage:
english_text = ""
gujarati_translation = translate_text(english_text)

print(f"English: {english_text}")
print(f"Gujarati: {gujarati_translation}")
```
